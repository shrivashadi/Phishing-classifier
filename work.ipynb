```python
import urllib.request
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00327/Training%20Dataset.arff"

try:
    urllib.request.urlretrieve(url, "phishing_dataset.arff")
    print("Download sussesfully!")
except Exception as e:
    print("Error:", e)
```

```python
from scipy.io import arff
import pandas as pd 
## Load Arff file
data, meta = arff.loadarff('phishing_dataset.arff')
```

```python
df = pd.DataFrame(data)
```

```python
## Bytes to int convert 
df = df.applymap(lambda x: int(x) if isinstance(x, bytes) else x)
print("converted successfully!")
print("Shape:", df.shape)
df.head()
```

```python
import os
os.makedirs('data', exist_ok=True)
df.to_csv('data/phishing.csv', index=False)
print("Data Saved Successfully!")
```

```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 

import warnings
warnings.filterwarnings("ignore")
```

```python
sns.countplot(x='Result', data=df)
plt.title('Phishing vs Legitimate Websites')
plt.xlabel('Result (-1=Phishing, 1=Legitimate)')
plt.ylabel('Count')
plt.show()
```

```python
df.hist(figsize=(20,15), bins=3)
plt.suptitle('Feature Distribution')
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(), annot=False, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
## TOP Correlated Features With Target
correlation = df.corr()['Result'].sort_values(ascending=False)
print("Top Features Correlated with Result:")
print(correlation)
```

```python
## Box Plot 
plt.figure(figsize=(20,8))
df.boxplot()
plt.xticks(rotation=90)
plt.title('Feature Boxplot')
plt.show()
```

```python
## Pairplot >> Importent Features
important_features = ['having_IP_Address', 'URL_Length', 
                      'SSLfinal_State', 'Domain_registeration_length',
                      'Request_URL', 'Result']

sns.pairplot(df[important_features], hue='Result')
plt.show()
```

```python
print("=== EDA Summary ===")
print(f"Total Samples: {df.shape[0]}")
print(f"Total Features: {df.shape[1]-1}")
print(f"\nPhishing Sites: {(df['Result']==-1).sum()}")
print(f"Legitimate Sites: {(df['Result']==1).sum()}")
print(f"\nClass Balance: {(df['Result']==-1).sum()/df.shape[0]*100:.2f}% Phishing")
```
